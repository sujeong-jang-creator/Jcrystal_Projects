## 강북구 장애인 인구 수

```python
data5 = pd.read_csv('dataset/202105.csv', encoding='cp949')
data5
```

```python
df5 = pd.DataFrame(data5)
df5
```

```python
sp = df5['통계시도명'] == '서울특별시'
sp_table = df5[sp]
sp_table
```

```python
kp = sp_table['통계시군구명'] == '강북구'
kp_table = sp_table[kp]
kp_table
```

```python
# 강북구 거주 장애인 총합
kp_total = kp_table['등록장애인수'].sum()
kp_total
```
